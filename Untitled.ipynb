# 2. checking if the treatment time is confounded with any other variables, with a decision tree classifier (random forest was tried but only limited increase in explanatory power), using ROC_AUC as an evaluation metric. We can see that treatmnet is not randomized, as it can be predicted with ~0.8ROCAUC
### Used tree model because nulls can be imputed with a dummy number, in this case -999

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import roc_auc_score
from sklearn.cluster import KMeans

clf = DecisionTreeClassifier(random_state=0,max_depth = 6,min_samples_leaf=30,
                             class_weight=None,criterion='entropy')
#clf = RandomForestClassifier(random_state=0,max_depth = 6,min_samples_split=30,class_weight='balanced',
#                            max_samples=0.5)

var = ['sex', 'age', 'other_drugs_1', 'other_drugs_2',
       'other_drugs_3', 'other_drugs_4', 'other_drugs_5', 'other_drugs_6',
       'other_drugs_7', 'other_drugs_8', 'diagnosis_1', 'diagnosis_2',
       'diagnosis_3', 'diagnosis_4', 'diagnosis_5', 'diagnosis_6',
       'diagnosis_7', 'diagnosis_8', 'diagnosis_9', 'diagnosis_10',
       'diagnosis_11', 'diagnosis_12', 'diagnosis_13', 'diagnosis_14',
       'diagnosis_15', 'lab_1', 'lab_2', 'lab_3', 'lab_4', 'lab_5', 'lab_6',
       'lab_7', 'lab_8', 'Diag_Score_1', 'Diag_Score_2',]
def scorer(clf,X,y):
    preds = clf.predict_proba(X)[:,1]
    return roc_auc_score(y,preds)
    
score = cross_val_score(clf, df[var].fillna(-999), df['Target'], cv=10,scoring=scorer)
print('CV ROC-auc-score:', score)
print('average ROC-auc-score:', np.mean(score),np.std(score)/10**.5)
```

# 3. Since treatment is not at random, build a test set that is properly randomized as test set to see how method performs

```python
#construct null hypothesis
x =[]
for i in range(200):
    s=roc_auc_score(df['Target'],np.random.uniform(0,1,size=len(df['Target'])))
    x += [s,]
```

```python
#first find things we should use for paired testing
impt_var = []
print ('Variables that are correlated with Treatment type')
for i in var:
    clf = DecisionTreeClassifier(random_state=0,max_depth = 3,min_samples_leaf=30,
                                 class_weight=None,criterion='entropy')

    def scorer(clf,X,y):
        preds = clf.predict_proba(X)[:,1]
        return roc_auc_score(y,preds)

    score = cross_val_score(clf, df[[i,]].fillna(-999), df['Target'], cv=10,scoring=scorer)
    if np.mean(score) > 0.5+2*np.std(x): 
        print (i,np.mean(score))
        impt_var += [i]
```

## Use PCA to dimension redcution followed by Kmeans to get 100 clusters, to get 100 groups for paired test

```python
pca = sklearn.decomposition.PCA(whiten=True)
preds = pca.fit_transform(df.fillna(df.median())[[x for x in impt_var if x not in var2]])
#use first 3 PCs, 98.5% variance
print ('PCA explained variance',np.cumsum(pca.explained_variance_ratio_)[:5])
for i in range(4):
    df['PC%i'%i]=preds[:,i]
kmeans = KMeans(150)

#use the 3 PCs + 3 most import var related to treatment type
X_kmeans = (df.fillna(df.median())[['PC0','PC1','PC2']+var2]/df.fillna(df.median())[['PC0','PC1','PC2']+var2].std())
kmeans.fit(X_kmeans)
df['Cluster']=kmeans.predict(X_kmeans).astype('str')
```

```python
df['index'] = df.index
d=df.groupby(['Cluster','Target']).apply(len).reset_index(drop=0)
d=pd.merge(d,df.groupby(['Cluster','Target'])['index'].apply(list).reset_index(drop=0))
d
```
